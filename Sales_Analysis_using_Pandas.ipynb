```python
import os
import pandas as pd
dir='.\SalesAnalysis\Sales_Data'
all_month_data=pd.DataFrame()
files=[file for file in os.listdir('./Sales_Data/')]

for file in files:
  df=pd.read_csv('./Sales_Data/'+file)
  all_month_data=pd.concat([all_month_data,df])

all_month_data.to_csv("all_month.csv")
```

```python
nan_df=all_month_data[all_month_data.isna().any(axis=1)]
nan_df.head()
```

```python
all_month_data=all_month_data.dropna(how='all')
```

```python
all_month_data.head()
all_month_data= all_month_data[all_month_data['Order Date'].str[0:2]!='Or']
all_month_data
```

```python
all_month_data['month']=all_month_data['Order Date'].str[0:2]
all_month_data['month']=all_month_data['month'].astype('int32')
all_month_data
```

```python
#convert column to numberic
all_month_data['Quantity Ordered']=pd.to_numeric(all_month_data['Quantity Ordered'])
all_month_data['Price Each']=pd.to_numeric(all_month_data['Price Each'])
```

```python
all_month_data['Sales']=all_month_data['Quantity Ordered']*all_month_data['Price Each']
all_month_data
```

```python
#remove duplicate order id
df=all_month_data[all_month_data['Order ID'].duplicated(keep=False)]
df
#here grouped with respect of "Order Id"
#1234 charger
#1234 mobile
#After this code, it can be like
# 1234 charger, mobile (grouped)
# 1234 charger, mobile (grouped)
#after removed duplicates
# we get single order with multiple product 
#1234 charger, mobile 
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x) )
df=df[['Grouped','Order ID']].drop_duplicates()
df.head()

#print(df.groupby('Order ID').sum())
```

```python
from itertools import combinations
from collections import Counter
count=Counter()

for row in df['Grouped']:
  #assume that you have list with commas
  row_list=row.split(',')
  count.update(combinations(row_list,3))
  #count.update(Counter(combinations(row,2)))

#print(count)
# print(count.most_common(10))

for key, value in count.most_common(10):
  print(key,value)
```

```python
# Find By Quantity
product_group=all_month_data.groupby('Product')
Quantity=product_group.sum()['Quantity Ordered']
#Get the "y" value only from the "x" Grouped data i.e if you group it by "product" get "y" from product grouped data
# x is product & y is Quantity
products=[product for product , df in all_month_data.groupby('Product')]


plt.bar(products,Quantity)
plt.xticks(products, rotation='vertical', size=8)
```
